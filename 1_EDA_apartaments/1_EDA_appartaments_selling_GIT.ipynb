## Предобработка данных

### Изменение названий столбцов 

```python
df.rename(columns={'cityCenters_nearest': 'city_centre_nearest', 'parks_around3000': 'parks_around_3_km', 'ponds_around3000': 'ponds_around_3_km'}, inplace=True)
```

### Обработка пропусков

В столбцах **total_images**, **last_price**, **total_area**, **first_day_exposition**, **rooms**, **floor**, **studio**, /
**open_plan** пропусков нет (23699 значений как и в строк в датафрейме).

Первый столбец с пропущенными значениями - **ceiling_height**. Высота потолков в квартирах варьируется в пределах от 2.5 до 5 метров в среднем ( но также есть вылетевшие значения, такие как 27.5 и 100 метров). Вследствие того, что высота потолков в квартире это величина, которая зачастую одинакова в квартирах с одинаковой планировкой, и варьируется в среднем в небольшом диапазоне, было принято решение заполнить пропуски в этом столбце медианным значением по данному столбцу (медиана для избежания влияния выбросов). 

```python
#  Построение гистограммы для определения разброса значений для переменной ceiling_height
df['ceiling_height'].hist(bins = 100, range = (0, 4))
```

```python
# заполнение пропусков столбца ceiling_height медианными значениями
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
```

```python
# гистограмма ceiling_height после обработки пропусков 
df['ceiling_height'].hist(bins = 100, range = (0, 4))
```

```python
## проверка на отсутствие пропущенных значений
df['ceiling_height'].isna().sum()
```

В столбце **floors_total** обнаружено 86 пропусков, что составляет всего 0.0036 от объема датафрейма. Было принято допущение заменить пропуски в данном столбце значениями из столбца **floor**, что не должно нарушить статистические характеристки вследствие малого объема пропусков данного вида.

```python
# количество пропусков в столбце  floors_total
df['floors_total'].isna().mean()
```

```python
df['floors_total'] = df['floors_total'].fillna(df['floor'])
```

```python
## проверка на отсутствие пропущенных значений
df['floors_total'].isna().sum()
```

```python
def title_parse_type(title):
    # функция для нахождения типа населенного пункта по нижнему регистру и его отделения от названия населенного пункта 
    res = ' '.join(word for word in title.split(' ') if word.islower())
    return res
def title_parse_name(title):
    # функция для нахождения названия населенного пункта по верхнему регистру и его отделения от типа населенного пункта 
    res = ' '.join(word for word in title.split(' ') if word[0].isupper())
    return res
```

Для удаления аномальных значений напишем функию, которая будет удалять данные, за пределами "усов" боксплота. Распределения наших данных аналогичны Пуассоновскому, поэтому было сделано предположение, удалять только то, что находится за пределами правого "уса". Однако мы уже видели что аномальные значения есть и слева, как например высота потолков в один метр. Для удаления выбросов были выбраны столбцы **ceiling_height, last_price, living_area, floor**, так как у многих квартир с аномальными значениями данных столбцов также будут аномальны значения столбцов **total_area**, **rooms**.

```python
def remove_outliers(col): 
    # функция для расчета межквартильного размаха и "усов", на выходе получаем массив булевых значений где False - аномальные значение. 
    Q1 = col.quantile(0.25)
    Q3 = col.quantile(0.75)
    IQR = Q3 - Q1    #межквартильный размах
    #df_out = col.loc[(col <= Q3 + 1.5 *IQR) & (col >= Q1 - 1.5 * IQR)]
    return (col <= Q3 + 1.5 *IQR) & (col >= Q1 - 1.5 * IQR)
```

Создадим новый датафрейм без аномальных значений 

```python
df_clear = df
df_clear = df_clear[remove_outliers(df['ceiling_height'])  &  remove_outliers(df['last_price']) & remove_outliers(df['living_area']) & remove_outliers(df['floor'])]
```
